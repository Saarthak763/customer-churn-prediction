# 03 — Modeling
Training and comparing Logistic Regression, Random Forest, and HistGradientBoosting.

```python
import sys
sys.path.append('../src')
from train import load_data, split_and_scale, train_logreg, train_random_forest, train_gradient_boosting, evaluate

X, y = load_data()
(X_train, X_test, X_train_scaled, X_test_scaled, y_train, y_test, scaler) = split_and_scale(X, y)
print(X_train.shape, X_test.shape)
```

## Baseline: Logistic Regression

```python
logreg = train_logreg(X_train_scaled, y_train)
_ = evaluate('Logistic Regression', logreg, X_test_scaled, y_test)
```

## Random Forest (tuned)

```python
rf = train_random_forest(X_train, y_train)
_ = evaluate('Random Forest', rf, X_test, y_test)
```

## HistGradientBoosting (tuned)

```python
hgb = train_gradient_boosting(X_train, y_train)
_ = evaluate('HistGradientBoosting', hgb, X_test, y_test)
```
