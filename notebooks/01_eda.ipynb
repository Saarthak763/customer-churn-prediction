# 01 — Exploratory Data Analysis
Customer Churn Prediction — Telco dataset (IBM sample)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
df = pd.read_csv('../data/raw/Telco-Customer-Churn.csv')
df.head()
```

## Shape, dtypes, missing values

```python
print(df.shape)
df.info()
df.isna().sum()
```

## Class balance — how imbalanced is churn?

```python
churn_rate = df['Churn'].value_counts(normalize=True)
print(churn_rate)

ax = df['Churn'].value_counts().plot(kind='bar', color=['#4C72B0', '#DD8452'])
ax.set_title('Churn vs Retained Customers')
ax.set_xlabel('Churn')
ax.set_ylabel('Count')
plt.savefig('../reports/figures/churn_balance.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Churn rate by contract type
Hypothesis: month-to-month customers churn far more than those on longer contracts.

```python
ct = pd.crosstab(df['Contract'], df['Churn'], normalize='index')
ct.plot(kind='bar', stacked=True, color=['#4C72B0', '#DD8452'])
plt.title('Churn Rate by Contract Type')
plt.ylabel('Proportion')
plt.savefig('../reports/figures/churn_by_contract.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Tenure distribution by churn

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='tenure', hue='Churn', bins=30, multiple='stack')
plt.title('Tenure Distribution by Churn Status')
plt.savefig('../reports/figures/tenure_by_churn.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Monthly charges vs churn

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Churn', y='MonthlyCharges')
plt.title('Monthly Charges by Churn Status')
plt.show()
```

## Internet service type vs churn

```python
ct2 = pd.crosstab(df['InternetService'], df['Churn'], normalize='index')
ct2.plot(kind='bar', stacked=True, color=['#4C72B0', '#DD8452'])
plt.title('Churn Rate by Internet Service Type')
plt.ylabel('Proportion')
plt.show()
```
