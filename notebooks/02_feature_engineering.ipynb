# 02 — Feature Engineering
Turning raw columns into model-ready, informative features.

```python
import pandas as pd
import sys
sys.path.append('../src')
from data_prep import load_raw, clean, engineer_features, encode

df = load_raw('../data/raw/Telco-Customer-Churn.csv')
df = clean(df)
df.head()
```

## New features
- `avg_monthly_spend` — TotalCharges / tenure (falls back to MonthlyCharges for new customers)
- `tenure_bucket` — 0-6mo / 6-12mo / 1-2yr / 2-4yr / 4-6yr
- `num_addon_services` — count of Yes across the 6 add-on service columns

```python
df_fe = engineer_features(df)
df_fe[['tenure', 'avg_monthly_spend', 'tenure_bucket', 'num_addon_services']].head(10)
```

## Does num_addon_services correlate with churn?

```python
df_fe.groupby('num_addon_services')['Churn'].mean()
```

## Final encoding check

```python
df_encoded = encode(df_fe)
print(df_encoded.shape)
df_encoded.dtypes.value_counts()
```
